```python
if age<=14:
  kids.isna().sum()
  kids.isnull().sum()
  cols = kids.columns
  data = kids[cols].values.flatten()
  s = pd.Series(data)
  s = s.str.strip()
  s = s.values.reshape(kids.shape)
  kids = pd.DataFrame(s, columns=kids.columns)
  kids = kids.fillna(0)
  kids.head()
  vals = kids.values
  symptoms = sy['Symptom'].unique()
  for i in range(len(symptoms)):
      vals[vals == symptoms[i]] = sy[sy['Symptom'] == symptoms[i]]['weight'].values[0]  
  d = pd.DataFrame(vals, columns=cols)
  d = d.replace('dischromic _patches', 0)
  d = d.replace('spotting_ urination',0)
  kids = d.replace('foul_smell_of urine',0)
  kids.head()
  (kids[cols] == 0).all()
  kids['Disease'].value_counts()
  kids['Disease'].unique()
  datak = kids.iloc[:,2:].values
  labelsk= kids['Disease'].values
  x_train, x_test, y_train, y_test = train_test_split(datak, labelsk)
  print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
  model = GaussianNB()
  model.fit(x_train, y_train)
  preds = model.predict(x_test)
  print(preds)
  conf_mat = confusion_matrix(y_test, preds)
  df_cm = pd.DataFrame(conf_mat, index=kids['Disease'].unique(), columns=kids['Disease'].unique())
  print('F1-score% =', f1_score(y_test, preds, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, preds)*100)
  sns.heatmap(df_cm)
elif 14<age<=55:
  adult.isna().sum()
  adult.isnull().sum()
  cols = adult.columns
  data = adult[cols].values.flatten()
  s = pd.Series(data)
  s = s.str.strip()
  s = s.values.reshape(adult.shape)
  adult = pd.DataFrame(s, columns=adult.columns)
  adult = adult.fillna(0)
  adult.head()
  vals = adult.values
  symptoms = sy['Symptom'].unique()
  for i in range(len(symptoms)):
      vals[vals == symptoms[i]] = sy[sy['Symptom'] == symptoms[i]]['weight'].values[0]  
  d = pd.DataFrame(vals, columns=cols)
  d = d.replace('dischromic _patches', 0)
  d = d.replace('spotting_ urination',0)
  adult= d.replace('foul_smell_of urine',0)
  adult.head()
  (adult[cols] == 0).all()
  adult['Disease'].value_counts()
  adult['Disease'].unique()
  dataa = adult.iloc[:,2:].values
  labelsa= adult['Disease'].values
  x_train, x_test, y_train, y_test = train_test_split(dataa, labelsa)
  print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
  model =GaussianNB()
  model.fit(x_train, y_train)
  preds = model.predict(x_test)
  print(preds)
  conf_mat = confusion_matrix(y_test, preds)
  df_cm = pd.DataFrame(conf_mat, index=adult['Disease'].unique(), columns=adult['Disease'].unique())
  print('F1-score% =', f1_score(y_test, preds, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, preds)*100)
  sns.heatmap(df_cm)
else:
  old.isna().sum()
  old.isnull().sum()
  cols = old.columns
  data = old[cols].values.flatten()
  s = pd.Series(data)
  s = s.str.strip()
  s = s.values.reshape(old.shape)
  old= pd.DataFrame(s, columns=old.columns)
  old= kids.fillna(0)
  old.head()
  vals = old.values
  symptoms = sy['Symptom'].unique()
  for i in range(len(symptoms)):
      vals[vals == symptoms[i]] = sy[sy['Symptom'] == symptoms[i]]['weight'].values[0]  
  d = pd.DataFrame(vals, columns=cols)
  d = d.replace('dischromic _patches', 0)
  d = d.replace('spotting_ urination',0)
  old= d.replace('foul_smell_of urine',0)
  old.head()
  (old[cols] == 0).all()
  old['Disease'].value_counts()
  old['Disease'].unique()
  datao = old.iloc[:,2:].values
  labelso= old['Disease'].values
  x_train, x_test, y_train, y_test = train_test_split(datao, labelso)
  print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
  model = GaussianNB()
  model.fit(x_train, y_train)
  preds = model.predict(x_test)
  print(preds)
  conf_mat = confusion_matrix(y_test, preds)
  df_cm = pd.DataFrame(conf_mat, index=old['Disease'].unique(), columns=old['Disease'].unique())
  print('F1-score% =', f1_score(y_test, preds, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, preds)*100)
  sns.heatmap(df_cm)
  
  
```
